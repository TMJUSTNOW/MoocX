## Compute distances

```python
import scipy.spatial.distance as dist
distMatrix = dist.pdist(movies.select_dtypes(include=['int64']), 'euclidean')
```

## Hierarchical clustering

```python
from scipy.cluster import hierarchy
clusterMovies = hierarchy.linkage(movies.select_dtypes(include=['int64']), method='ward')
```

## Plot the dendrogram

```python
print hierarchy.dendrogram(clusterMovies)
```

## Assign points to clusters  
The `fcluster` function from `scipy.cluster.hierarchy` is not that convinent to use, ie. you can simply specifying the number of clusters you want.    
I have experimented with various `distance` values to get 10 clusters.  

```python
clusterGroups = hierarchy.fcluster(clusterMovies, 10.4, 'distance')
```

## Now let's figure out what the clusters are like.
Let's use the `crosstab` function to compute the percentage of movies in each genre and cluster

```python
tableAction =  pd.crosstab(movies['Action'], clusterGroups)
print tableAction.loc[1]/tableAction.sum()
```

```python
tableRomance =  pd.crosstab(movies['Romance'], clusterGroups)
print tableRomance.loc[1]/tableRomance.sum()
```
