```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import FileLinks
```

```python
from glob import glob
from termcolor import colored as c, cprint
from pprint import pprint
import numpy as np
import pickle
```

```python
data_files = list(np.sort(glob('dumps/xys_*.dump.pkl')))
pprint(len(data_files))
```

```python
data_files = list(np.sort(glob('dumps/xys_*.dump.pkl')))
# pprint(data_files)

title = 'Electron Configuration\nDuring Annealing (WIP)'

filenames = []
for i, file_path in enumerate(data_files[::20]):
    with open(file_path, 'rb') as f:
        xys = pickle.load(f)
        
    plt.figure(figsize=(4, 4))
    plt.scatter(xys[:, 0], xys[:, 1], color="#23aaff", edgecolor='none')
    plt.xlim(-60e-3, 60e-3)
    plt.ylim(-60e-3, 60e-3)
    plt.title(title)
    plt.xlabel('x coordinate (mm)')
    plt.ylabel('y coordinate (mm)')
    filename = "figures/Electron Configuration/" + title.replace('\n', ' ') + '_' + str(1000 + i)[-3:] +'.png'
    # plt.show()
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    filenames.append(filename)
```
