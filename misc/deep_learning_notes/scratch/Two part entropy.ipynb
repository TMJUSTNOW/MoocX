```python
import matplotlib.pyplot as plt
import numpy as np

p0s = np.arange(0, 1, 0.01)

def entropy (*ps):
	return - sum(ps * np.log2(ps))

etas = map(lambda p0: entropy(p0, 1-p0), p0s)
```

```python
plt.plot(p0s, etas)
plt.ylim(-0.5, 1.5)
plt.title("Entropy Vs. P_0 in a two-part system");
```

```python
entropy(0.5, 0.5)
```
