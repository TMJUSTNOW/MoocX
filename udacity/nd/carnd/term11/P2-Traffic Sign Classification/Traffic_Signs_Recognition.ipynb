Implemented Min-Max scaling in the `normalize_img()` function to a range of `a=-1.0` and `b=1.0`. 
After scaling, the values of the pixels in the input data should range from -1 to 1.
Min-Max Scaling:
$
X'=a+{\frac {\left(X-X_{\min }\right)\left(b-a\right)}{X_{\max }-X_{\min }}}
$

````python
### Preprocess the data here.
### Feel free to use as many code cells as needed.
from sklearn.preprocessing import LabelBinarizer
#http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelBinarizer.html

# Knowledge gained from TensorFlow Lab
# Implemented Min-Max scaling for the image data
# Transform features by scaling each feature to a given range.
def normalize_img(img):
    x_min = 0
    x_max = 255
    a = -1.0
    b = 1.0
    return a + (img - x_min) * (b - a) / (x_max - x_min)

X_train = normalize_img(X_train)  
X_test = normalize_img(X_test)

# Implemented one-hot encoding
encoder = LabelBinarizer()
encoder.fit(y_train)
y_train = encoder.transform(y_train)
y_test = encoder.transform(y_test)

# Change to float32, so it can be multiplied against the features in TensorFlow, which are float32
y_train = y_train.astype(np.float32)
y_test = y_test.astype(np.float32)
````

Tensorflow Recap
- tf.placeholder() returns a tensor that gets its value from data passed to the tf.session.run() function, allowing us
  to set the input right before the session runs.
- tf.truncated_normal() function returns a tensor with random values from a normal distribution whose magnitude is no 
  more than 2 standard deviations from the mean.

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.

# [IMP] CNN for Visual Recognition: http://cs231n.github.io/convolutional-networks/#fc
import tensorflow as tf

# Input image size
image_size = X_train[0].shape[0]
num_channels = X_train[0].shape[2]

# Filter size
kernel_size = 5
depth1 = 32
depth2 = 64
num_hidden = 1024

graph = tf.Graph()
with graph.as_default():
    features = tf.placeholder(tf.float32, [None, image_size, image_size, num_channels])
    labels = tf.placeholder(tf.float32, [None, n_classes])
    keep_prob = tf.placeholder(tf.float32)
    beta = tf.placeholder(tf.float32)
    learning_rate = tf.placeholder(tf.float32)
    
    # Imp: Depth doesn't change during pooling operation
    # The goal of training a neural network is to modify weights and biases to best predict the labels.
    
    weights =  {
        'layer1': tf.Variable(tf.truncated_normal(shape = [kernel_size, kernel_size, num_channels, depth1], 
                                                  stddev = 0.1)),
        'layer2': tf.Variable(tf.truncated_normal(shape = [kernel_size, kernel_size, depth1, depth2], 
                                                  stddev = 0.1)),
        # '4' here because we have 2 max pooling with stride of 2
        'layer3': tf.Variable(tf.truncated_normal(shape = [image_size // 4 * image_size // 4 * depth2, num_hidden], 
                                                  stddev = 0.1)),
        'layer4': tf.Variable(tf.truncated_normal(shape = [num_hidden, n_classes], stddev = 0.1))
    } 

    ''' Introduction to TensorFlow Quiz 13
    Weights are already helping prevent the model from getting stuck, 
    we don't need to randomize the bias. Let's use the simplest solution, setting the bias to 0.'''
    
    biases = {
        'layer1': tf.Variable(tf.zeros([depth1])),
        'layer2': tf.Variable(tf.zeros([depth2])),
        'layer3': tf.Variable(tf.zeros([num_hidden])),
        'layer4': tf.Variable(tf.zeros([n_classes]))
    }
    
    ''' CNN Lesson 19
    -ksize and strides : [batch, height , width, channels] 
   
    -tf.nn.max_pool() function performs max pooling with the ksize parameter as the size of the filter and 
    the strides parameter as the length of the stride. 2x2 filters with a stride of 2x2 are used commonly.
   
    -If we use 'SAME' padding with a stride of 1 the output width and height are the same as input.
    zero padding is performed to match the size
   
    -DNN Quiz 6 [Learned the implemenation too]
    Rectified linear unit (ReLU) is type of activation function that is defined as f(x) = max(0, x). 
    The function returns 0 if x is negative, otherwise it returns x. 
    
    - CNN Lesson 17
    tf.nn.conv2d() function to compute the convolution with weight as the filter and [1, 1, 1, 1] for the strides. 
    TensorFlow uses a stride for each input dimension, [batch, input_height, input_width, input_channels]. 
    Generally we set the stride for batch and input_channels(first and fourth element in the strides array) to be 1.
    input in my case is features.'''
    
    # Hidden Layer with ReLU activation function [2nd line]
    # convolution -> activation -> pooling [Repeat the rounds :D]
    #http://stackoverflow.com/questions/37674306/what-is-the-difference-between-same-and-valid-padding-in-tf-nn-max-pool-of-t
    # out_height = ceil(float(in_height) / float(strides[1]) out_width = ceil(float(in_width) / float(strides[2]))
    # Convolution Layer 1, input size: 32 x 32 x 1 output size: 32 x 32 x 32 [H*W*D]
    conv = tf.nn.conv2d(features, weights['layer1'], strides = [1, 1, 1, 1], padding = 'SAME')
    hidden = tf.nn.relu(conv + biases['layer1'])
    pool = tf.nn.max_pool(hidden, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'SAME')
    # Pooling Layer 1, input size: 32 x 32 x 32 output size: 16 x 16 x 32 [H*W*D]
    
    # Convolution Layer 2, input size: 16 x 16 x 32 output size: 16 x 16 x 64
    conv = tf.nn.conv2d(pool, weights['layer2'], [1, 1, 1, 1], padding = 'SAME')
    hidden = tf.nn.relu(conv + biases['layer2'])
    pool = tf.nn.max_pool(hidden, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'SAME')
    # Pooling Layer 2, input size: 16 x 16 x 64 output size: 8 x 8 x 64 [H*W*D]
    
    # Fully Connected Layer 1, output size: 1 x 1024
    shape = pool.get_shape().as_list()
    reshape = tf.reshape(pool, shape = [-1, shape[1] * shape[2] * shape[3]])
    hidden = tf.nn.relu(tf.matmul(reshape, weights['layer3']) + biases['layer3'])
    hidden = tf.nn.dropout(hidden, keep_prob)
    
    # Fully Connected Layer 2, output size: 1 x 43 
    # [Output layer with linear activation: W X + b = Y]
    # Linear Classifer is used and the scores are changed to probabilities later. 
    # Intro to TensorFlow Quiz 13, DNN Lesson 11
    
    logits = tf.matmul(hidden, weights['layer4']) + biases['layer4']

    '''Now let's prevent the network from over optimising on the training set.
       Regularization: Applying artificial constraints on the network that implicitly reduces 
       the number of free parameters while not making it more difficult to optimize.
       L2 Regularization from Lesson 15 CNN
       L2 norm of our weights * beta (small constant)'''
    
    regul = beta * (tf.nn.l2_loss(weights['layer1']) + 
                     tf.nn.l2_loss(weights['layer2']) +
                     tf.nn.l2_loss(weights['layer3']) +
                     tf.nn.l2_loss(weights['layer4']) )
    
    # Defined loss and  Optimizer DNN Lesson 11
    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, labels) + regul)
    optimizer = tf.train.AdamOptimizer(learning_rate).minimize(loss)
    
    # Probability
    # Introduction to TensorFlow Lesson 12
    # Converting the scores (logits) to probabilities using softmax function
    prediction = tf.nn.softmax(logits)
    
    # Evaluation op phase :D
    #tf.nn.top_k : https://www.tensorflow.org/api_docs/python/nn/evaluation
    #It returns two tensors: 1)top K value 2) indices of these values in the original tensor.
    top_five = tf.nn.top_k(prediction, 5)
    
    # Check if predictions are correct
    correct_prediction = tf.equal(tf.argmax(prediction, 1), tf.argmax(labels, 1))
    # Calculate Accuracy
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    # Save Model
    saver = tf.train.Saver()
````

````python
### Train your model here.
### Feel free to use as many code cells as needed.

# Maximum epoch number
epochs = 25

# Batch size
batch_size = 128

# Training data size
train_data_size = X_train.shape[0]

# Feed_dict for validation data
feed_dict_valid = {features: X_valid, labels: y_valid, keep_prob: 1.0}

# Feed_dict for test data
feed_dict_test = {features: X_test, labels: y_test, keep_prob: 1.0}

with tf.Session(graph = graph) as session:
    tf.global_variables_initializer().run()
    #tf.initialize_all_variables().run()
    for epoch in range(1, epochs + 1):
        # Shuffle the indices for the trainning data
        ind = np.arange(0, train_data_size)
        np.random.shuffle(ind)
        X_train_shuffle = X_train[ind]
        y_train_shuffle = y_train[ind]
        # Go through all the batches
        # 'train_data_size - batch_size' to ignore the residual samples at end of each epoch
        for batch_ind in range(0, train_data_size - batch_size, batch_size):
            batch_data = X_train_shuffle[batch_ind : (batch_ind + batch_size), :, :, :]
            batch_labels = y_train_shuffle[batch_ind : (batch_ind + batch_size), :]
            feed_dict_train = { features: batch_data, 
                                labels: batch_labels, 
                                keep_prob: 0.5,
                                beta: 1e-6,
                                learning_rate: 0.001 }
            _, l, acc = session.run([optimizer, loss, accuracy], feed_dict = feed_dict_train)
        print("Epoch: %d\tloss: %f\tAccuracy: %f" %(epoch, l, acc))
        valid_acc = accuracy.eval(feed_dict = feed_dict_valid)
        print("Validation accuracy:\t%f" %valid_acc)
        # If both accuracy better than 0.99, stop running
        if(valid_acc > 0.99 and acc > 0.99):
            break
    print("Test accuracy:\t%f" %accuracy.eval(feed_dict = feed_dict_test))
    # save_model = saver.save(session, "Final-check1")
    saver.save(session, "Final-check1")
    # print("Model saved as %s" %save_model)
````
